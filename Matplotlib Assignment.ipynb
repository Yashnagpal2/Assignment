Matplotlib is a popular Python library used for creating static, animated, or interactive visualizations in 2D and 3D. It is widely used in data analysis, scientific research, and various fields of data visualization. Matplotlib provides a wide range of plotting options and customization capabilities, making it a versatile tool for creating high-quality plots and charts.

Matplotlib is used for tasks such as:

Data Visualization: Matplotlib allows you to create a wide variety of plots to visualize data, making it easier to understand and analyze datasets.

Exploratory Data Analysis: It is a valuable tool for exploring data during the initial stages of data analysis, helping you identify patterns, trends, and outliers.

Educational Purposes: Matplotlib is also used in educational contexts to teach data visualization and programming.

Line Plot: A basic line plot is used to visualize data points connected by lines, which is often used to represent trends or changes over a continuous variable.

Scatter Plot: Scatter plots are used to represent individual data points as dots on a 2D plane, which is helpful for exploring relationships between two variables.

Q2:What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a type of data visualization that is used to display individual data points on a two-dimensional coordinate system. In a scatter plot, each data point is represented as a dot or marker at the intersection of two numerical variables, typically one variable on the x-axis and another on the y-axis. It is a useful tool for visually exploring the relationship or correlation between two variables.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
plt.title("Scatter Plot")
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It is useful when you want to display multiple plots or charts in a single figure, allowing you to arrange and customize their positions and sizes. Each subplot can display different data or visualizations, and you can control their layout using row and column specifications.

```python
data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
]

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2)

# Plot data in each subplot
for i, (x, y) in enumerate(data):
    row, col = i // 2, i % 2  # Calculate the row and column for the current subplot
    ax = axes[row, col]  # Get the current subplot
    ax.plot(x, y)  # Plot the data in the current subplot
    ax.set_title(f'Line Plot {i + 1}')  # Set subplot title

# Add space between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

A bar plot, also known as a bar chart or bar graph, is a data visualization that represents categorical data with rectangular bars. The length (or height) of each bar is proportional to the value it represents. Bar plots are used to display and compare the values of different categories or groups
Comparison: Bar plots are useful for comparing data across categories and identifying differences in values.

Categorical Data: They are well-suited for visualizing and representing categorical data with distinct groups or items.

Frequency Distribution: Bar plots display the frequency distribution of categorical data, showing how often each category occurs.

Part-to-Whole Relationships: Stacked bar plots allow representing the composition of a whole in terms of its parts.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a statistical data visualization that provides a summary of the distribution of a dataset. It is used for various purposes in data analysis and visualization:

Summary of Data Distribution: Box plots provide a concise summary of the distribution of data, including measures such as the median, quartiles, and potential outliers.

Identifying Central Tendency: Box plots display the median as the central line inside the box, making it easy to identify the dataset's central tendency.

Detection of Skewness and Symmetry: The shape of the box and the position of the median in the box plot can indicate whether the data is skewed or symmetrically distributed.

Outlier Detection: Box plots include whiskers that extend to show potential outliers in the data, helping to identify data points that fall far outside the expected range.

```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)

# Add labels to the x-axis
plt.xticks([1, 2], ['Box 1', 'Box 2'])

# Add a title
plt.title('Box Plot for Box 1 and Box 2')

# Show the plot
plt.show()
```
