Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots that you can create using Seaborn, along with their common use cases:

Scatter Plot (sns.scatterplot):

Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a dot on a 2D plane, making it suitable for exploring correlations, clusters, or patterns.
Histogram (sns.histplot):

Use: Histograms are used to show the distribution of a single variable. They group data into bins and display the frequency or density of data points within each bin. Histograms are helpful for understanding data distributions and identifying patterns.
Box Plot (sns.boxplot):

Use: Box plots display the summary statistics of a dataset, including the median, quartiles, and potential outliers. They are used for visualizing and comparing the distributions of different categories or groups, making them suitable for data exploration and statistical analysis.
Heatmap (sns.heatmap):

Use: Heatmaps are used to represent data in a 2D matrix format, where each cell's color encodes a value. Heatmaps are commonly used to visualize correlations between variables, display confusion matrices, or represent various data relationships in a visually intuitive way.
Bar Plot (sns.barplot):

Use: Bar plots are used to visualize and compare categorical data. They display the relationship between a categorical variable and a numeric variable. Bar plots are effective for showing comparisons, ranking categories, and summarizing data in an easily interpretable manner.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot( x =fmri.timepoint , y = fmri.signal,data = fmri,style = "region" ,hue = "event")
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot =flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d', linewidths=0.5)
```
