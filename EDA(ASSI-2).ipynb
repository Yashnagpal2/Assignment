#### Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
min = df['Price'].min()
max = df['Price'].max()
range = df['Price'].max() - df['Price'].min()
```

```python
print("The min is:" ,min)
print("The max is:",max)
print("The range is:",range)
```

#### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=df, palette='viridis')
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Flight Price')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

#### Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price', data=df, color='skyblue')
plt.title('Boxplot of Flight Prices')
plt.xlabel('Flight Price')
plt.show()
```

In the boxplot, potential outliers are data points that fall beyond the whiskers of the box. Outliers may impact your analysis in the following ways:

Influence on Descriptive Statistics:

Outliers can significantly affect measures like the mean and standard deviation, making them less representative of the majority of the data.
Skewing Distributions:

Outliers can skew the distribution of prices, making it appear more spread out than it actually is.
Impact on Models:

In predictive modeling, outliers can influence the performance of models. For instance, they may disproportionately affect regression coefficients.
Data Quality and Anomalies:

Outliers may indicate data quality issues or anomalies in the data collection process. It's essential to investigate the reasons behind outliers.
Consideration in Analysis:

Depending on the nature of your analysis, you may choose to handle outliers by removing them, transforming the data, or applying robust statistical techniques.

#### Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.

Understanding and handling missing values is crucial for a robust analysis, as missing data can impact the results in several ways:

Biased Analysis: Missing data can introduce bias into the analysis if the missing values are not random. For example, if certain data is missing systematically, it may lead to inaccurate conclusions.

Reduced Sample Size: Missing values reduce the effective sample size for analysis. This can impact the statistical power and reliability of your results.

Incomplete Insights: If important variables have a high percentage of missing values, it may limit your ability to draw meaningful insights from those variables.

Model Performance: In machine learning, models might not handle missing values well. Imputation or removal strategies may be needed to use these models effectively.

Data Quality Assessment: Identifying missing values is also a part of assessing overall data quality. It prompts consideration of data collection methods and potential issues in the data collection process.
